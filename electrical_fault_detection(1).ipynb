```python
data['fault_status'] = ((data['G'] + data['A'] + data['B'] + data['C']) > 0).astype(int)
```

```python
conditions = [
    # LLLG - Three phase to ground (A, B, C, G all 1)
    (data['A'] == 1) & (data['B'] == 1) & (data['C'] == 1) & (data['G'] == 1),

    # LLL - Three phase (A, B, C are 1, G is 0)
    (data['A'] == 1) & (data['B'] == 1) & (data['C'] == 1) & (data['G'] == 0),

    # LLG - Two lines to ground (e.g., A&B&G, and the third line is 0)
    (data['A'] == 1) & (data['B'] == 1) & (data['G'] == 1) & (data['C'] == 0),
    (data['B'] == 1) & (data['C'] == 1) & (data['G'] == 1) & (data['A'] == 0),
    (data['A'] == 1) & (data['C'] == 1) & (data['G'] == 1) & (data['B'] == 0),

    # LG - One line to ground (e.g., A&G, and other lines are 0)
    (data['A'] == 1) & (data['G'] == 1) & (data['B'] == 0) & (data['C'] == 0),
    (data['B'] == 1) & (data['G'] == 1) & (data['A'] == 0) & (data['C'] == 0),
    (data['C'] == 1) & (data['G'] == 1) & (data['A'] == 0) & (data['B'] == 0),

    # LL - Two lines (no ground, and the third line is 0)
    (data['A'] == 1) & (data['B'] == 1) & (data['C'] == 0) & (data['G'] == 0),
    (data['B'] == 1) & (data['C'] == 1) & (data['A'] == 0) & (data['G'] == 0),
    (data['A'] == 1) & (data['C'] == 1) & (data['B'] == 0) & (data['G'] == 0)
]

choices = [
    'LLLG - three phase to ground',
    'LLL - three phase',
    'LLG - A and B to ground',
    'LLG - B and C to ground',
    'LLG - A and C to ground',
    'LG - A to Ground',
    'LG - B to Ground',
    'LG - C to Ground',
    'LL - A to B',
    'LL - B to C',
    'LL - A to C'
]

data['fault_type'] = np.select(conditions, choices, default='No Fault')
```

```python
def calculate_rms(series):
    return np.sqrt(np.mean(series**2))

# Calculate RMS for current and voltage columns
data['Ia_rms'] = data.groupby('fault_type')['Ia'].transform(calculate_rms)
data['Ib_rms'] = data.groupby('fault_type')['Ib'].transform(calculate_rms)
data['Ic_rms'] = data.groupby('fault_type')['Ic'].transform(calculate_rms)
data['Va_rms'] = data.groupby('fault_type')['Va'].transform(calculate_rms)
data['Vb_rms'] = data.groupby('fault_type')['Vb'].transform(calculate_rms)
data['Vc_rms'] = data.groupby('fault_type')['Vc'].transform(calculate_rms)

print("New RMS features added to the DataFrame:")
print(data[['Ia_rms', 'Ib_rms', 'Ic_rms', 'Va_rms', 'Vb_rms', 'Vc_rms']].head())
```

```python
columns_for_rms_correlation = ['fault_status', 'Ia_rms', 'Ib_rms', 'Ic_rms', 'Va_rms', 'Vb_rms', 'Vc_rms']
correlation_matrix_rms = data[columns_for_rms_correlation].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_rms, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Fault Status vs. RMS Current and Voltage')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 1. Setup the figure and 3D axis
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# 2. Define your axes (using RMS for a cleaner view)
x = data['Ia_rms']
y = data['Ib_rms']
z = data['Ic_rms']

# 3. Create the scatter plot
# We use 'fault_type' for the color (hue)
# Note: You may need to encode your 'fault_type' to integers for the 'c' parameter
# or use a loop to plot each group separately for a legend.
categories = data['fault_type'].unique()
colors = sns.color_palette("hls", len(categories))

for i, cat in enumerate(categories):
    subset = data[data['fault_type'] == cat]
    ax.scatter(subset['Ia_rms'], subset['Ib_rms'], subset['Ic_rms'],
               label=cat, s=20, alpha=0.6, color=colors[i])

# 4. Labeling
ax.set_xlabel('Current Ia (RMS)')
ax.set_ylabel('Current Ib (RMS)')
ax.set_zlabel('Current Ic (RMS)')
ax.set_title('3D State Space: Current Distribution by Fault Type')
ax.legend(title="Fault Type", loc='best')

plt.tight_layout()
plt.show()
```
